```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/postings.csv')

print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
```

```python
df.info()
```

```python
# الأعمدة المهمة بس
cols = [
    'job_id', 'title', 'company_name', 'location',
    'formatted_work_type', 'formatted_experience_level',
    'normalized_salary', 'pay_period', 'remote_allowed'
]

df_clean = df[cols].copy()

print("Shape:", df_clean.shape)
df_clean.head()
```

```python
# تحميل ملفات الـ skills
df_job_skills = pd.read_csv('../data/job_skills.csv')
df_skills_map = pd.read_csv('../data/skills.csv')

print("Job Skills Shape:", df_job_skills.shape)
print("\nJob Skills Sample:")
print(df_job_skills.head())

print("\nSkills Map Sample:")
print(df_skills_map.head())
```

```python
df_job_skills_named = df_job_skills.merge(df_skills_map, on='skill_abr', how='left')

df_full = df_clean.merge(df_job_skills_named, on='job_id', how='left')

print("Shape:", df_full.shape)
print("\nSample:")
df_full.head(10)
```

```python

df_full.to_csv('../output/jobs_clean.csv', index=False)
df_clean.to_csv('../output/jobs_main.csv', index=False)

print("✅ Saved")
print(f"jobs_clean.csv  → {df_full.shape[0]:,} rows")
print(f"jobs_main.csv   → {df_clean.shape[0]:,} rows")
```
